```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('brain_stroke.csv')
df
```

```python
a=[]
for i in df['age']:
  if i<1:
    i=i*10
    a.append(i)
  else:
    a.append(i)
```

```python
df['age']=a
```

```python
df['age']=df['age'].astype(int)
```

```python
df['age']=df['age'].replace(0,np.nan)
```

```python
df.dropna(inplace=True)
```

```python
df.drop('ever_married',axis=1,inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['gender']=label.fit_transform(df['gender'])
df['work_type']=label.fit_transform(df['work_type'])
df['Residence_type']=label.fit_transform(df['Residence_type'])
df['smoking_status']=label.fit_transform(df['smoking_status'])
```

```python
df.head()
```

```python
x=df.drop('stroke',axis=1)
y=df['stroke']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(xtrain,ytrain)
rf.score(xtest,ytest)
```

```python
rf.predict([[0,45,1,0,0,0,95.02,31.8,3]])
```

```python
import pickle
file=open('brainstroke.pkl','wb')
pickle.dump(rf,file)
```
